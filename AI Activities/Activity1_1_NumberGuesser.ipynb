# 🎯 Number Guesser – Basic Random Number Prediction
This beginner-friendly game introduces students to simple Python logic and interaction through a text-based number guessing game.

```python
# Run this cell to start the game!
import random

secret_number = random.randint(1, 100)
attempts = 0

print("I have picked a number between 1 and 100. Try to guess it!")

while True:
    guess = input("Enter your guess: ")
    if not guess.isdigit():
        print("❌ Please enter a valid number.")
        continue

    guess = int(guess)
    attempts += 1

    if guess == secret_number:
        print(f"✅ Correct! You guessed it in {attempts} tries.")
        break
    elif guess < secret_number:
        print("🔼 Too low. Try a higher number.")
    else:
        print("🔽 Too high. Try a lower number.")
```

## 🏦 Real-World Example: Two-Factor Authentication (2FA)
In the real world, this logic is used for security. Instead of guessing a number for fun, a system generates a random **One-Time Password (OTP)** that the user must match to log in.

```python
import random

# 1. System generates a secret 4-digit code
stored_otp = random.randint(1000, 9999)
max_attempts = 3
current_try = 0

print(f"[DEBUG] A code was sent to your phone. (Hint: {stored_otp})")

while current_try < max_attempts:
    user_input = input(f"Attempt {current_try + 1}/{max_attempts} - Enter 4-digit code: ")

    if not user_input.isdigit() or len(user_input) != 4:
        print("Invalid format! Please enter 4 digits.")
        continue

    if int(user_input) == stored_otp:
        print("🔓 Access Granted! Welcome back.")
        break
    else:
        print("❌ Incorrect code.")
        current_try += 1

if current_try == max_attempts:
    print("🚫 Account Locked. Too many failed attempts.")
```
